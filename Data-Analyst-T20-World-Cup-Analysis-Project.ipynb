```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

Match Summary 


```python
#import json
import pandas as pd
with open(r'D:\Data Analyst T20 project\t20_json_files\t20_wc_match_results.json') as f:
    data = json.load(f)
df_matches = pd.DataFrame(data[0]['matchSummary'])
df_matches.head()
```

```python
df_matches.shape            #rows and columns
```

```python
#rename coulmn name scorecard  to match_id
df_matches.rename({'scorecard':'match_id'}, axis=1,inplace=True)

# Normalize match names: lowercase & strip spaces
df_matches["team1"] = df_matches["team1"].str.strip().str.lower()
df_matches["team2"] = df_matches["team2"].str.strip().str.lower()

df_matches.head()
```

```python
match_id_dict = {}

for index, row in df_matches.iterrows():
    key1 = row['team1'] + ' vs ' + row['team2']
    key2 = row['team2'] + ' vs ' + row['team1']
    
    match_id_dict[key1] = row["match_id"]
    match_id_dict[key2] = row["match_id"]

match_id_dict
```

Batting Summary

```python
with open(r'D:\Data Analyst T20 project\t20_json_files\t20_wc_batting_summary.json') as f:
    data = json.load(f)

    all_records = []    

    for record in data:
        all_records.extend(record['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
#creating a new column out/not_out

df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not out')
df_batting.head(15)
```

```python
 #removing special characters from batsmanName column

df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Â',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('†',''))  
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ã³',''))  
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ã©',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ã¨',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ãª',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ã¼',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ã¤',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ã¶',''))
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('Ã±',''))

# Normalize match names: lowercase & strip spaces
df_batting["match"] = df_batting["match"].str.strip().str.lower()

df_batting.head(15)
```

```python
df_batting["match"] = df_batting["match"].str.replace("  ", " ").str.strip()  # Remove double spaces
df_batting["match"] = df_batting["match"].str.replace("Vs", "vs")  # Ensure lowercase 'vs'
```

```python
df_batting["match_id"] = df_batting["match"].map(match_id_dict)
df_batting.head()
```

```python
# Check which matches couldn't be mapped

unmapped_matches = df_batting[df_batting["match_id"].isna()]["match"].unique()
print("Unmapped Matches:", unmapped_matches)
```
